```python
import numpy as np
from sklearn import metrics
import matplotlib.pyplot as plt
import pandas as pd
from scipy.interpolate import make_interp_spline, BSpline
```

```python
df = pd.read_csv('zHydrogen.csv', usecols=[0,1])
df.dropna(inplace = True)

df_new = df.astype(float)

x = np.array(df_new['energy'].to_list()) #to convert data in a list to npArray
y = np.array(df_new['length'].to_list())
```

```python
#w/ smoothing
plt.figure(figsize=(5, 3), dpi=300)

plt.xlabel(r'Distance ($\AA{}$)')
plt.ylabel(r"Energy (Eh)")

# 300 represents number of points to make between x.min and x.max
xnew = np.linspace(x.min(), x.max(), 300)  

spl = make_interp_spline(x, y, k=3)
y_smooth = spl(xnew)

#create smooth line chart 
plt.plot(xnew, y_smooth)

plt.xticks(x, rotation='vertical', fontsize = 5)
# Pad margins so that markers don't get clipped by the axes
plt.margins(0.1)
# Tweak spacing to prevent clipping of tick-labels
plt.subplots_adjust(bottom=0.15)

plt.show()
```
